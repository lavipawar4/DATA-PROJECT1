```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load your transactional data into a pandas DataFrame
transaction_data = pd.read_csv('Hackathon_Working_Data.csv')

# Load your transactional data into a pandas DataFrame
transaction_data = pd.read_csv('Hackathon_Working_Data.csv')

# Explore basic statistics and information about the dataset
print(transaction_data.info())
print(transaction_data.describe())

# Exploratory Data Analysis (EDA)
# Visualize purchasing patterns and customer preferences
# Example: Distribution of Bill amounts
plt.figure(figsize=(8, 6))
sns.histplot(transaction_data['BILL_AMT'], bins=20, kde=True)
plt.title('Distribution of Bill Amount')
plt.xlabel('Bill Amount')
plt.ylabel('Frequency')
plt.show()

```

```python
# Analyze customer demographics
# Example: Count plot of purchsing day
plt.figure(figsize=(10, 6))
sns.countplot(x='DAY', data=transaction_data, order=transaction_data['DAY'].value_counts().index)
plt.title('Distribution of Purchase Days')
plt.xlabel('Purchase Days')
plt.ylabel('Count')
plt.show()
```

```python
# Identify popular product categories
# Example: Pie chart of product brand category distribution
plt.figure(figsize=(10, 8))
transaction_data['BRD'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Product Brand Categories')
plt.ylabel('')
plt.show()
```
